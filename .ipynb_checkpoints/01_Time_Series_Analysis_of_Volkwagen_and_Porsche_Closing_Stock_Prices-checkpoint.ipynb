# PROJECT DESCRIPTION:
In this project, I'll be analyzing the closing stock prices of Volkwagen (VW) and Porsche in the periods leading up to, and following the official announcements of the buyout of Porsche by Volkswagen.


I'll examine the effects that the buyout had on the closing stock prices, and how they changed in the aftermath of the following official announcements:
1. First official announcement, made on 2009-12-09. The day VW announced they owned 49.9% of Porsche.
2. Second official announcement, made on 2012-07-05. The day VW announced they acquired full ownership (51.1%) of Porsche.

The entire period of study has been divided into three distinct intervals; the split points of which are the dates of the official announcements:
- Interval 1: Before the first official announcement.
- Interval 2: Between the first and second official announcements.
- Interval 3: After the second official announcement.

The analysis will focus on exploring what the effects the buyout had on the following across the 3 intervals:
- The trend in the stock,
- The correlation between the stock,
- How the best fitting models for each stock changed, and
- Volatility of the stocks.

To ensure that all changes in values are stock specific, we need a market benchmark. The ideal choice for this would be some other automobile producer selling stocks on the German market.  
Bayerische Motoren Werke AG (BMW) is chosen to be the benchmark. Therefore the stock prices of BMW in the period of study will be used as the baseline of the analysis.

The objective of this project is to explore and analyse the specific events that have been defined rather than an attempt to forecast future stock prices of VW and Porsche.



### Dataset Description:
- The data is scrapped from yahoo finance using Python package `yfinance`.
- It raw data contains records from as far back as 1996 up till the date it was scrapped, however the analysis will be limited to between period 2009-04-05, and 2014-01-01.
- The three(3) companies, whose equitites were recorded from the German Stock Market, to be examined and their respective tickers are:
    - PAH3.DE Porsche Automobil Holding SE
    - VOW3.DE: Volkswagen AG
    - BMW.DE: Bayerische Motoren Werke AG. (Chosen benchmark for the analysis)
- The currency of the stock prices is the EURO, â‚¬.

# DATA PREPARATION AND PRE-PROCESSING

### Importing the Relevant Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.tsa.stattools as sts
from pmdarima.arima import auto_arima
import yfinance
import warnings

warnings.filterwarnings("ignore")
sns.set()

%matplotlib inline
```

### Importing the Data
Download the data from Yahoo finance:

```python
raw_data = yfinance.download(tickers="VOW3.DE, PAH3.DE, BMW.DE", interval="1d", group_by='ticker',
                              auto_adjust=True, treads=True)
```

```python
# create copy of data in a new dataframe.
df = raw_data.copy()
```

```python
print(df.shape)
df.head(3)
```

### Defining Key Dates
Defining the important dates for this specific business case:

```python
# Starting Date
start_date = "2009-04-05" 

# First Official Announcement - 49.9%  - Day VW announced they owned 49.9% of Porsche
ann_1 = "2009-12-09" 

# Second Official Announcement - 51.1% - Day VW announced they full ownership of Porsche (51.1%)
ann_2 = "2012-07-05" 

#Ending Date
end_date = "2014-01-01"
```

### Pre-processing the Data

```python
# Extracting Closing Prices
df['vol'] = df['VOW3.DE'].Close
df['por'] = df['PAH3.DE'].Close
df['bmw'] = df['BMW.DE'].Close

# Creating Returns
df['ret_vol'] = df['vol'].pct_change(1).mul(100)
df['ret_por'] = df['por'].pct_change(1).mul(100)
df['ret_bmw'] = df['bmw'].pct_change(1).mul(100)

# Creating Squared Returns - required for analysing volatility
df['sq_vol'] = df.ret_vol.mul(df.ret_vol)
df['sq_por'] = df.ret_por.mul(df.ret_por)
df['sq_bmw'] = df.ret_bmw.mul(df.ret_bmw)

# Extracting Volume - number of purchases and sales each day
df['q_vol'] = df['VOW3.DE'].Volume
df['q_por'] = df['PAH3.DE'].Volume
df['q_bmw'] = df['BMW.DE'].Volume
```

Assigning the Frequency of the time series data and Filling NA Values

```python
# Frequency set to business days since stock prices trading can only be done in those days.
df = df.asfreq('b')

# Using backfill method for filling missing values.
df = df.fillna(method='bfill')
```

```python
# Removing Surplus Data
del df['VOW3.DE']
del df['PAH3.DE']
del df['BMW.DE']

df.head()
```

###### Save dataframe to external csv file

```python
columns = [col[0] for col in df.columns]


df.to_csv("vol_por_bmw.csv", header=columns)
```

# FINDING THE BEST FITTING MODELS
We've seen how the stock prices of the two companies, the market benchmark, and the correlation between them changed as a result of the buyout.  
Next, I'll explore how the best fitting model that described each company's stock prices changed across the three intervals.

An event like a buyout is expected to change the market significantly, but what happens to the best fitting models?   
The best fitting models are supposed to capture the current trend in the market, but if the trend shifts then so should they. This is exactly what I hope to determine; how the best fitting models changed as the market did.  
I'll realise this by finding the best fitting models for each stock in the three interval and compare each.

The best fitting models for each company's stock prices across the the three intervals will be found using the `auto_arima()`.  <br />
The `auto_arima()` finds the best fitting models by taking in different parameters which include, but are not limited to: 
    
>`exogenous` - exogenous variables,  
`seasonal` - Whether to fit a seasonal ARIMA (default=True), 
`m`-seasonal cycle length/frequency,  
`max_p` - maximum AR components,  
`max_q` - maximum MA components,  
`max_d` - maximum AR integrations,  
`stationary` - The stationarity of the time series (default = False),  
`information_criterion` - information criterion to use, 'aic', 'bic', 'hqic', etc.
`alpha` - significance level (default is 5%), 


With these parameters setting constraints, it fits the provided data to models with different combinations of these parameters and chooses the best model, i.e the one with the lowest information criteria (AIC), and statistically significant coefficients.

### Stationarity Tests
Before commencing the analysis, it is important to note whether or not the stock prices of the companies are stationary or not. Performing modelling under the assumption of stationarity without first confirming this could lead to the models producing incorrect / misleading results.

I'll begin by performing a test to determine whether the stock prices of the two company's and the market benchmark are stationary or not.  
This will be done by using the Augmented Dickey-Fuller test.  

The null hypothesis of the Augmented Dickey-Fuller test assumes non-stationarity. Consequently, if the test statistic of the test is less than the critical value, then the null hypothesis is rejected and we conclude that the data is stationary.  
Alternatively, we can compare the p-value of the test with the chosen significance level. If the p-value is less than the significance level we can also conclude that the data is stationary, and vice versa.

###### stationarity test for VW

```python
adfuller_result_vol = sts.adfuller(df.vol)
adfuller_result_vol
```

_interpretation:_
- The first line is the test statistic from running the test.  
Comparing it to the critical value at 5% significance level (-2.86198) we see that it is greater, therefore VW's stock prices data is **NOT** stationary.  
- We also see that the p-value (second line) is significantly greater than the significance level (5% or 0.05). This is further proof that it is not stationary.

```python
# Is the test statistic less than the critical value at 5% level of significance?
adfuller_result_vol[0] < adfuller_result_vol[4]['5%']
```

OR

```python
# Is the p-value less than the 5% significance level?
adfuller_result_vol[1] < 0.05
```

###### stationarity test for Porsche

```python
adfuller_result_por = sts.adfuller(df.por)
adfuller_result_por
```

```python
# Is the test statistic less than the critical value at 5% level of significance?
adfuller_result_por[0] < adfuller_result_por[4]['5%']
```

- Porsche's stock prices data is **NOT** stationary. 

###### stationarity test for BMW

```python
adfuller_result_bmw = sts.adfuller(df.bmw)
adfuller_result_bmw
```

```python
adfuller_result_bmw = sts.adfuller(df.bmw)
adfuller_result_bmw
```

```python
# Is the test statistic less than the critical value at 5% level of significance?
adfuller_result_bmw[0] < adfuller_result_bmw[4]['5%']
```

- BMW's stock prices data is **NOT** stationary. 

---
Now it has been determined that none of the stock prices data are stationary, we'll need to explicitly specify this in the parameters of the  `auto-ARIMA` models so that the models know to include performing differencing during the process of finding the best model.

### Best Fitting Models for Volkswagen

###### Best Fitting Models for Volkswagen: Interval 1

```python
mod_interval_1_vol = auto_arima(df.vol[start_date:ann_1], exogenous=df[['por','bmw']][start_date:ann_1],
                             stationary=False, seasonal=True, m=5, max_p=5, max_q=5, 
                             information_criterion='aic', n_jobs=-1, trace=True)

mod_interval_1_vol.summary()
```

_interpreing the auto_arima result: SARIMAX(p,d,q)(P,D,Q)x_
- "Best model: ARIMA(1,0,0)(0,0,0)[5] intercept".
    - The values for `d`, `q`, `P`, `D`, `Q` are all 0, so we know that there is no integration order, MA component, and no seasonality.
    - Therefore, we are left with an **AR(1)** model with a single lag and two exogenous variables as the best fit.   
- All coefficients apart from the intercept are significant, so this is indeed a good fit.  
i.e. their p-values (`P>|z|`) are less than 0.05.
- The estimate for the AR(1) value  (`ar.L1`) is very close to 1. This means we are sticking really close to the value of the last period with vey little deviation.

###### Best Fitting Models for Volkswagen: Interval 2

```python
mod_interval_2_vol = auto_arima(df.vol[ann_1:ann_2], exogenous=df[['por','bmw']][ann_1:ann_2],
                             stationary=False, seasonal=True, m=5, max_p=5, max_q=5, 
                             n_jobs=-1, trace=True)

mod_interval_2_vol.summary()
```

_interpretation:_
- Here, "Best model:  ARIMA(0,1,0)(0,0,0)[5]"  
    - The values for `p`, `q`, `P`, `D`, `Q` are all 0, so we know that there is AR component, MA component, no seasonality, and no intercept.  
    We are left with a model with only an integration order.
    - This is a special case of the ARIMA models: `ARIMAX(0,1,0)`, a _random walk model_. 
- It is intriguing, for the best fitting model for VW stock prices in the interval between announcements, to be a "random walk" model.
- This model assumes that at each point in time, the stock prices merely take a random step away from the its last recorded position, and do not rely on past values. The mean value of the steps taken is zero.
- The rises in the coefficients of the exogenous variables from the previuos interval, (Porsche: 0.40->0.7908 ~50% rise, and BMW: 0.6919->1.5487 ~44% rise) reveal that they have more weight in estimations than in the previous interval.

To ascertain that the model is a good fit, I'll plot the residual plots, and review them to get more insights, but before that, I'll also plot the stock prices of the companies in this interval so we can see it more closely:

- The "Standardized residual" plot clearly resembles a white noise process:
    - The residuals fluctuate around a mean of zero.
    - They have uniform variance.
    - There is no discernable pattern in the plot, suggesting no autocorrelation of values.  
    
The residuals of random walk processes are assumed to be white noise. This is clearly the case here.
- The "Histogram plus estimated density" plot suggests a normal distribution with mean of zero.
- The "Normal Q-Q" plot shows that most of the residuals falls along the straight line. This suggests normality of the residual.
- The Correlogram (ACF plot) show that the residual errors are not autocorrelated.

All of these validate the assumptions of the model, and reveal that the model infact explains most of the variability of the data, thus proving that it is a good fit.

The process will be repeated for interval 2 and interval 3.

### Measuring Model Accuracy with Mean Absolute Percentage Error (MAPE)
I'll quantify the accuracy of the predictions by measuring the Mean Absolute Percentage Error (MAPE).  
MAPE measures how accurate the forecast system is.

_Mathematical Formula for MAPE:_  
$$ \mbox M = \frac{1}{n} \sum_{t-1}^{n} \mid\frac{A_t - F_t}{A_t}\mid$$
 Where:
* $n$ is the number of fitted points
* $A_t$ is the actual value
* $F_t$ is the forecast value

```python
# Function to calculate MAPE
def mape(forecast, actual):
    percent_err = np.mean(np.abs(actual - forecast)/np.abs(actual))
    return f"Model has a percentage error of {percent_err:.2%}"
```

# VOLATILITY
Volatility is simply a measure of risk or uncertainty associated with a security. It measures the dispersion of a security's returns from its mean or the average return.  
Volatility is often quantified by the standard deviation or variance of price returns.

Let's examine volatility for VW and Porsche prices across the three intervals. Usually, we expect some major shifts surrounding such events

_interpretation:_
