# IMPORTING THE PACKAGES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from arch import arch_model
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
sns.set()

%matplotlib inline
```

# DATA PREPARATION

### Importing the Data

```python
raw_data = pd.read_csv('vol_por_bmw.csv', index_col='Date')

df = raw_data.copy()

# Convert index to DatetimeIndex object.
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

### Defining Key Dates
Defining the important dates for this specific business case:

```python
# Starting Date
start_date = "2009-04-05" 

# First Official Announcement - 49.9%  - Day VW announced they owned 49.9% of Porsche
ann_1 = "2009-12-09" 

# Second Official Announcement - 51.1% - Day VW announced they full ownership of Porsche (51.1%)
ann_2 = "2012-07-05" 

#Ending Date
end_date = "2014-01-01"
```

# VOLATILITY
<!-- refers to some numeric measurement of the uncertainty which depends on the volume of changes in valuation. It  -->
Volatility is simply a measure of risk or uncertainty associated with a security. It measures the dispersion of a security's returns from its mean or the average return.  
Volatility is often quantified by the standard deviation or variance of price returns.

Let's examine volatility for VW and Porsche prices across the three intervals. Usually, we expect some major shifts surrounding such events.

### Volatility: Volkswagen

```python
df['sq_vol'][start_date:end_date].plot(figsize=(20,5), color="blue")

plt.axvline(x=ann_1, ymax=1, ls="-.", c="black")
plt.axvline(x=ann_2, ymax=1, ls="-.", c="black")
plt.annotate(f"""First Announcement
{ann_1}""", 
            xy=(ann_1, 230), xycoords='data',
            bbox=dict(boxstyle="round", fc="none", ec="gray"),
            xytext=(80, 0), textcoords='offset points', ha='center',
            arrowprops=dict(arrowstyle="simple"))
plt.annotate(f"""Second Announcement
{ann_2}""", 
            xy=(ann_2, 230), xycoords='data',
            bbox=dict(boxstyle="round", fc="none", ec="gray"),
            xytext=(80, 0), textcoords='offset points', ha='center',
            arrowprops=dict(arrowstyle="simple"));
```

_interpretation:_
- VW has the highest volatility before any of the announcements. 
- Fascinatingly, the stock becomes relatively stable following each of the announcements, while experiencing instability in the periods leading up to the announcements.  
Such behaviour can be attributed to the rumours spreading across the markets about a deal on the horizon. However, since Porsche had tried to purchase VW a few years back, many people were not certain about how it would play out.

### Volatility: Porsche

```python
df['sq_por'][start_date:end_date].plot(figsize=(20,5), color="green")

plt.axvline(x=ann_1, ymax=1, ls="-.", c="black")
plt.axvline(x=ann_2, ymax=1, ls="-.", c="black")
plt.annotate(f"""First Announcement
{ann_1}""", 
            xy=(ann_1, 300), xycoords='data',
            bbox=dict(boxstyle="round", fc="none", ec="gray"),
            xytext=(80, 0), textcoords='offset points', ha='center',
            arrowprops=dict(arrowstyle="simple"))
plt.annotate(f"""Second Announcement
{ann_2}""", 
            xy=(ann_2, 300), xycoords='data',
            bbox=dict(boxstyle="round", fc="none", ec="gray"),
            xytext=(80, 0), textcoords='offset points', ha='center',
            arrowprops=dict(arrowstyle="simple"));
```

_Interpretation:_
- After initially having high volatility, the stock eventually becomes stable in the period leading to, and following the first official announcement.
- The stocks become fully stable after the second announcement, i.e after the completion of the buyout.  
This could be attributed to the fact that its stock value, as well as VW's are now being viewed as a single entity, and as VW grows as a market leader, the speculation and thus uncertainty surrounding the value of Porsche's stock prices greatly diminishes.

# VOLATILITY MODELLING
Here, I'll model the volatility trends of VW and Porsche's stock prices for each interval.  
It has been mathematically proven that no higher GARCH model outperforms the GARCH(1,1) when it comes to variance of market returns, so I'll use it to examine how the variance and residual terms' coefficients (_omega_, _alpha_, and _beta_) evolve from interval to interval.  

`GARCH(1,1)` model is described mathematically by the mean and variance equations:  
_Mean Equation:_ $r_t = \mu_t + \varepsilon_t$  <br />
_Variance Equation:_ $Var(y_t|y_{t-1}) = \Omega + \alpha_1\varepsilon^2_{t-1} + \beta^2_{t-1}\sigma^2_{t-1}$

- $\mu_t$ mean,
- $\varepsilon_t$ residual term,
- $\Omega$ numerical constant,
- $\alpha_1$ is the coefficient of the squared residual from the past period. (also known as, reaction parameter),
- $\beta$ is the coefficient of the conditional variance from the last period. (also known as, persistence parameter),
- $\varepsilon^2_{t-1}$ squared residual from the past period,
- $\sigma^2_{t-1}$ is the conditional variance from the last period.

Conditional volatility is the volatility of some random variable given some extra information. Conditional volatility in GARCH models is conditioned on past values of itself and of model errors.  

Creating and fitting the `GARCH(1,1)` models for each interval:

### Modelling Volkswagen's Volatility

###### Volkswagen Volatility: Interval 1

```python
model_garch_interval_1_vol = arch_model(df.ret_vol[start_date:ann_1], 
                             mean="Constant", vol="GARCH", p=1, q=1)
results_garch_interval_1_vol = model_garch_interval_1_vol.fit(update_freq=5)

results_garch_interval_1_vol.summary()
```

_interpretation:_
- The relatively high value of `alpha[1]` ($\alpha$), 0.1540, tells us that the market prices for VW in this interval is 'jumpy' or 'nervous' (as we saw in its volatility plot [above](#Volatility:-Volkswagen)).
- `beta[1]` ($\beta$), has a relatively low value. Its p-value of 1.0, suggests full multicollinearity.  
This means that the volatility is not expected to persist for long periods into the future since the decay is faster.  
Indeed, we see this in its volatility plot that after this interval the prices become relatively stable.
- The high ($\alpha$) value combined with the low ($\beta$) value tells us that the volatility in this interval has a high volatility.

```python
# Plotting the standardized residuals and conditional volatility
results_garch_interval_1_vol.plot();
```

###### Volkswagen Volatility: Interval 2

```python
model_garch_interval_2_vol = arch_model(df.ret_vol[ann_1:ann_2], mean="Constant",
                                    vol="GARCH", p=1, q=1)
results_garch_interval_2_vol = model_garch_interval_2_vol.fit(update_freq=5)

results_garch_interval_2_vol.summary()
```

_interpretation:_
- p-value of `beta[1]` is equal to 0.0. This is the exact opposite of what we saw in the results of the previous interval. 
- This means the autocorrelation in the conditional variance is significant, and hardly dies off based on the coefficient value.
- The low `alpha[1]` value coupled with the high `beta[1]` value suggests that the prices in this interval are relatively stable.

```python
# Plotting the standardized residuals and conditional volatility
results_garch_interval_2_vol.plot();
```

- This plot shows the autocorrelation present in the conditional variance.

###### Volkswagen Volatility: Interval 3

```python
model_garch_interval_3_vol = arch_model(df.ret_vol[ann_2:end_date], mean="Constant", 
                                    vol="GARCH", p=1, q=1)
results_garch_interval_3_vol = model_garch_interval_3_vol.fit(update_freq=5)

results_garch_interval_3_vol.summary()
```

_interpretation:_
- The coefficient for the constant `omega` and `alpha[1]` are non-significant.  
This implies that there is some autocorrelation and volatility which fits the _volatility clustering_ feature (i.e, periods of high volatility are followed by periods of high volatility, and periods of low volatility are followed by periods of low volatility). 
- However, since the entire interval is much more stable, even the `beta[1]` coeff has dropped by almost one-third. It is also non-significant.

```python
# Plotting the standardized residuals and conditional volatility
results_garch_interval_3_vol.plot();
```

- The conditional volatility plot confirms the summary of the results.

### Modelling Porsche's Volatility

###### Porsche Volatility: Interval 1

```python
df['sq_por'][start_date:end_date].plot(figsize=(20,5), color="green")

plt.axvline(x=ann_1, ymax=1, ls="-.", c="black")
plt.axvline(x=ann_2, ymax=1, ls="-.", c="black")
plt.annotate(f"""First Announcement
{ann_1}""", 
            xy=(ann_1, 300), xycoords='data',
            bbox=dict(boxstyle="round", fc="none", ec="gray"),
            xytext=(80, 0), textcoords='offset points', ha='center',
            arrowprops=dict(arrowstyle="simple"))
plt.annotate(f"""Second Announcement
{ann_2}""", 
            xy=(ann_2, 300), xycoords='data',
            bbox=dict(boxstyle="round", fc="none", ec="gray"),
            xytext=(80, 0), textcoords='offset points', ha='center',
            arrowprops=dict(arrowstyle="simple"));
```

```python
model_garch_interval_1_por = arch_model(df.ret_por[start_date:ann_1], 
                             mean="Constant", vol="GARCH", p=1, q=1)
results_garch_interval_1_por = model_garch_interval_1_por.fit(update_freq=5)

results_garch_interval_1_por.summary()
```

_interpretation:_
- The `mu` is significant which suggests there is some constant trend in the returns values.
- The coefficient `alpha[1]`, has a value ~ 0.0000; this means that the immediate impact of any shock to the stock prices is extremely low.
- `beta[1]` has a very high value, indicating that the duration of any impact to the stock prices persists for longer.
- Taken together, the very low impact of shocks coupled with its persistence for longer periods could mean a continuous decrease in volatility.

```python
results_garch_interval_1_por.plot();
```

- We observe a continuous decay in the conditional volatility.

###### Porsche Volatility: Interval 2

```python
model_garch_interval_2_por = arch_model(df.ret_por[ann_1:ann_2], mean="Constant",
                                    vol="GARCH", p=1, q=1)
results_garch_interval_2_por = model_garch_interval_2_por.fit(update_freq=5)

results_garch_interval_2_por.summary()
```

_interpretation:_
- The `mu` here is also significant which suggests there is some constant trend in the returns values.
- p-value of `beta[1]` is extremely close to 0. This means the autocorrelation in the conditional variance is significant, and hardly dies off based on the coefficient value.
- The low `alpha[1]` value coupled with the high `beta[1]` value suggests that the prices in this interval are relatively stable.

```python
results_garch_interval_2_por.plot();
```

- This plot shows the autocorrelation present in the conditional variance.

###### Porsche Volatility: Interval 3

```python
model_garch_interval_3_por = arch_model(df.ret_por[ann_2:end_date], mean="Constant", 
                                    vol="GARCH", p=1, q=1)
results_garch_interval_3_por = model_garch_interval_3_por.fit(update_freq=5)

results_garch_interval_3_por.summary()
```

_interpretation:_
- The relatively low `alpha[1]` value coupled with the relatively high `beta[1]` value suggests that the prices in this interval are relatively very stable, as we saw in Porsche's volatility plot [above](#Volatility:-Porsche)

```python
results_garch_interval_3_por.plot();
```

- There is some autocorrelation present in the conditional volatility

# SUMMARY
Overall, we see really different volatility trends in all 3 intervals for VW, and Porsche which is quite intriguing. Both started off being highly volatile in interval 1, and eventually became relatively stable in interval 3.  

The purchase of Porsche has led to the death of many rumours and provided VW with the larger share of the market. This in return has resulted in much lower volatility after each announcement making VW stocks a lot more appealing prospect.
